```python
df = df.drop(columns=['Id'])
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
from sklearn import preprocessing

encoder = preprocessing.OneHotEncoder(drop='first')
encoded_data = encoder.fit_transform(df[categorical_cols])
encoded_col_names = encoder.get_feature_names_out(categorical_cols)
one_hot_encoded_frame = pd.DataFrame(encoded_data.toarray(), columns=encoded_col_names, index=df.index)

df = pd.concat([df.drop(columns=categorical_cols), one_hot_encoded_frame], axis=1)
```

```python
y= df['SalePrice']

y_log = np.log(df['SalePrice'])

X = df.drop(columns=['SalePrice'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X, y_log, test_size=0.2, random_state=99)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])

X_train_log_scaled = X_train_log.copy()
X_test_log_scaled = X_test_log.copy()

X_train_log_scaled[numeric_cols] = scaler.fit_transform(X_train_log[numeric_cols])
X_test_log_scaled[numeric_cols] = scaler.transform(X_test_log[numeric_cols])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error

linear = LinearRegression()
linear.fit(X_train_scaled, y_train)

y_train_pred = linear.predict(X_train_scaled)
y_test_pred = linear.predict(X_test_scaled)

train_RMSE = root_mean_squared_error(y_train, y_train_pred)
test_RMSE = root_mean_squared_error(y_test, y_test_pred)

train_R2 = r2_score(y_train, y_train_pred)
test_R2 = r2_score(y_test, y_test_pred)

n_test = len(y_test)
p = X_train_scaled.shape[1]
test_adj_R2 = 1 - (1-test_R2)*(n_test-1)/(n_test-p-1)

print("Train RMSE:", train_RMSE)
print("Test RMSE:", test_RMSE)
print("Train R^2:", train_R2)
print("Test R^2:", test_R2)
print("Test Adjusted R^2:", test_adj_R2)
```

```python
lin_log = LinearRegression()
lin_log.fit(X_train_log_scaled, y_train_log)

y_train_pred_log = lin_log.predict(X_train_log_scaled)
y_test_pred_log = lin_log.predict(X_test_log_scaled)

train_RMSE_log = root_mean_squared_error(y_train_log, y_train_pred_log)
test_RMSE_log = root_mean_squared_error(y_test_log, y_test_pred_log)

train_R2_log = r2_score(y_train_log, y_train_pred_log)
test_R2_log = r2_score(y_test_log, y_test_pred_log)
test_adj_R2_log = 1 - (1-test_R2_log)*(n_test-1)/(n_test-p-1)

print("Train RMSE (log):", train_RMSE_log)
print("Test RMSE (log):", test_RMSE_log)
print("Train R^2:", train_R2_log)
print("Test R^2:", test_R2_log)
print("Test Adjusted R^2:", test_adj_R2_log)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

ridge_cv_scores_log = {}
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

for a in alphas:
    model = Ridge(alpha=a)
    cv_mse = cross_val_score(
        model, X_train_log_scaled, y_train_log,
        scoring='neg_mean_squared_error', cv=5)
    ridge_cv_scores_log[a] = cv_mse.mean()

best_alpha_ridge_log = max(ridge_cv_scores_log, key=ridge_cv_scores_log.get)
print("Best alpha:", best_alpha_ridge_log)

ridge_log = Ridge(alpha=best_alpha_ridge_log)
ridge_log.fit(X_train_log_scaled, y_train_log)

y_train_pred = ridge_log.predict(X_train_log_scaled)
y_test_pred = ridge_log.predict(X_test_log_scaled)

test_RMSE_log = root_mean_squared_error(y_test_log, y_test_pred)
test_R2_log = r2_score(y_test_log, y_test_pred)
test_adj_R2_log = 1 - (1-test_R2_log)*(n_test-1)/(n_test-p-1)

print("Train RMSE (log):", train_RMSE_log)
print("Test RMSE (log):", test_RMSE_log)
print("Train R^2:", train_R2_log)
print("Test R^2:", test_R2_log)
print("Test Adjusted R^2:", test_adj_R2_log)
```

```python
from sklearn.linear_model import Lasso

lasso_cv_scores_log = {}

for a in alphas:
    model = Lasso(alpha=a, max_iter=5000)
    cv_mse = cross_val_score(
        model, X_train_log_scaled, y_train_log,
        scoring='neg_mean_squared_error', cv=5
    )
    lasso_cv_scores_log[a] = cv_mse.mean()

best_alpha_lasso_log = max(lasso_cv_scores_log, key=lasso_cv_scores_log.get)
print("Best alpha:", best_alpha_lasso_log)

lasso_log = Lasso(alpha=best_alpha_lasso_log, max_iter=5000)
lasso_log.fit(X_train_log_scaled, y_train_log)

y_test_pred_log = lasso_log.predict(X_test_log_scaled)

test_RMSE_log = root_mean_squared_error(y_test_log, y_test_pred_log)
test_R2_log = r2_score(y_test_log, y_test_pred_log)
test_adj_R2_log = 1 - (1-test_R2_log)*(n_test-1)/(n_test-p-1)

print("Train RMSE (log):", train_RMSE_log)
print("Test RMSE (log):", test_RMSE_log)
print("Train R^2:", train_R2_log)
print("Test R^2:", test_R2_log)
print("Test Adjusted R^2:", test_adj_R2_log)
```

```python
coefs = pd.Series(lasso_log.coef_, index=X.columns)
selected = coefs[coefs != 0].sort_values(ascending=False)

print("Selected features with coefficients:")
print(selected)

print("\nNumber selected:", (coefs != 0).sum(), "out of", len(coefs))
```

```python
from sklearn.linear_model import ElasticNet

elastic_cv_scores_log = {}

alphas = [0.001, 0.01, 0.1, 1, 10, 100]
l1_ratios = [0.1, 0.5, 0.9]

for a in alphas:
    for l1 in l1_ratios:
        model = ElasticNet(alpha=a, l1_ratio=l1, max_iter=5000)
        cv_mse = cross_val_score(
            model, X_train_log_scaled, y_train_log,
            scoring='neg_mean_squared_error', cv=5
        )
        elastic_cv_scores_log[(a, l1)] = cv_mse.mean()

best_alpha_elastic_log, best_l1_ratio_elastic_log = max(elastic_cv_scores_log, key=elastic_cv_scores_log.get)
print("Best alpha:", best_alpha_elastic_log)
print("Best l1_ratio:", best_l1_ratio_elastic_log)

elastic_log = ElasticNet(alpha=best_alpha_elastic_log, l1_ratio=best_l1_ratio_elastic_log, max_iter=5000)
elastic_log.fit(X_train_log_scaled, y_train_log)

y_train_pred_elastic_log = elastic_log.predict(X_train_log_scaled)
y_test_pred_elastic_log = elastic_log.predict(X_test_log_scaled)

train_RMSE_elastic_log = root_mean_squared_error(y_train_log, y_train_pred_elastic_log)
test_RMSE_elastic_log = root_mean_squared_error(y_test_log, y_test_pred_elastic_log)
train_R2_elastic_log = r2_score(y_train_log, y_train_pred_elastic_log)
test_R2_elastic_log = r2_score(y_test_log, y_test_pred_elastic_log)
test_adj_R2_elastic_log = 1 - (1-test_R2_elastic_log)*(n_test-1)/(n_test-p-1)

print("Train RMSE (log):", train_RMSE_elastic_log)
print("Test RMSE (log):", test_RMSE_elastic_log)
print("Train R^2:", train_R2_elastic_log)
print("Test R^2:", test_R2_elastic_log)
print("Test Adjusted R^2:", test_adj_R2_elastic_log)
```
